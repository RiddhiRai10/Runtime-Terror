```python
def timing(dep):
    dep = dep.split(':')
    dep = int(dep[0])
    if (dep >= 6 and dep < 12):
        return 'Morning'
    elif (dep >= 12 and dep < 17):
        return 'Noon'
    elif (dep >= 17 and dep < 20):
        return 'Evening'
    else:
        return 'Night'
    
train_df['Timing'] = train_ban['Timing'].apply(timing)   
test_df['Timing'] = test_ban['Timing'].apply(timing) 

train_df['Timing'] = train_ban['Timing'].apply(timing)   
test_df['Timing'] = test_ban['Timing'].apply(timing) 
```
